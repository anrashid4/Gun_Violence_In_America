```python
# Import depdendencies 
import pandas as pd 
import numpy as np
```

```python
# Change certain values in column where necessary
gun['weapons_obtained_legally'].replace({"Kelley passed federal criminal background checks; the US Air Force failed to provide information on his criminal history to the FBI": "Yes", "\nYes": "Yes"}, inplace=True)
```

```python
# View the changes 
gun["weapons_obtained_legally"].value_counts()
```
